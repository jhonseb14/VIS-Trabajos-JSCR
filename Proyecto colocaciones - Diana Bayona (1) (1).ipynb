### ¿Qué?

Para esto se tienen 2 variables categóricas, la primera equivale a la "Línea de Producción" en donde hay 15 categorías: sostenimiento, siembras, compra de animales, microcrédito, producción, infraestructura y adecuo de tierras, comercialización, otras actividades, maquinaria y equipo, comercialización, consolidación de pasivos, actividades rurales y servicios de apoyo, estas dos últimas se desagregan por capital de trabajo e inversión; la segunda variable equivale al tipo de productor, en donde se tienen las categorías pequeño, mediano y grande. Adicionalmente, existe un atributo categórico que corresponde al ID del crédito otrogado, junto con un atributo ordenado secuencial que corresponde al valor de la colocación. Finalmente, los datos que se tienen son de tipo **tabla** ya que está cuenta con 4 atributos y 488466 items que representan la información de la colocación de los créditos agropecuarios en el 2020, así mismo los datos que se tienen son estáticos.

```python
sns.catplot(data = df,
            kind = "bar",
            estimator = sum,
            x = "Colocacion",
            y = "Linea de Produccion",
            col = "Tipo Productor",
            palette = 'magma',
            ci = None)
```

```python
df.groupby(['Linea de Produccion', 'Tipo Productor']).agg({"Colocacion" : ["count", "mean", "sum"]})
```

### ¿Por qué?

En en sus acciones la visualización **analiza y presenta** ya que el propósito es mostrar las colocaciones de créditos por línea de producción y tipo de productor, siendo los productores grandes los que presentan las colocaciones de créditos de mayor valor en las lineas de producción de comercialización y servicios de apoyo de capital de trabajo. Así mismo, el gráfico **consulta y compara** como acción ya que a partir de este se puede comparar los valores de colocación por línea de producción y por cada uno de los tipos de productor. Por ejemplo, para los productores pequeños las colocaciones de crédito son para líneas de producción de siembras y compras de animales, mientras que para el productor mediano los créditos son para líneas de producción de sostenimiento y compra de animales, finalmente para el productor grande las lineas de producción de comercialización y servicios de apoyo de capital de trabajo presentan las colocaciones de crédito de mayor valor.

Con respecto al objetivo, considero que este es **Tendencias** para todos los datos y con respecto a los atributos el objetivo es **distribución** dado que por medio del gráfico se pueden identificar las características de los datos y como se distribuyen estos ya que en el modismo se presentan las líneas de producción con mayor valor de colocación por tipo de productor. Igualmente, se puede evidenciar que la colocación de los créditos tiene un mayor valor para los productores grandes, mientras que para los productores pequeños y medianos no se presenta una gran diferencia en los valores de colocación.

**Tarea**

La tarea y objetivo que se quiere lograr por medio de este gráfico es poder identificar las líneas de producción que tienen un mayor valor en la colocación de los créditos por tipo de productor, de esta manera Finagro quien es la entidad encargada de otorgar estas colocaciones de créditos puede ofrecer diferentes créditos según el tipo de productor y la línea de producción.
